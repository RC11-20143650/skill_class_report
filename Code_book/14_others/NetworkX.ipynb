```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class Film:
    def __init__(self, title, director, cast):
        self.title = title
        self.director = director
        self.cast = cast
```

```python
with open("netflix_titles.csv") as csvfile:
    reader = csv.reader(csvfile)
    print(next(reader))
    print(next(reader))
```

```python
with open("netflix_titles.csv") as csvfile:
    reader = csv.reader(csvfile)
    rows = 0
    for r in reader:
        rows +=1
    print(rows)
```

```python
Films = []
with open("netflix_titles.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        title = row['title']
        director = row['director']
        cast = row['cast'].split(', ')
        Films.append(Film(title, director, cast))
```

```python
print(Films[20].cast)
print(Films[20].title)
```

```python
for c in Films[0].cast:
    print(c)
```

```python
FilmGraph = nx.Graph()
```

```python
filmSubset = Films[0:50]
```

```python
for f in filmSubset:
    FilmGraph.add_node(f, title = unicode(f.title[:10], "utf-8"))
```

```python
for f1 in filmSubset:
    for f2 in filmSubset:
        if not f1 == f2:
            if f1.director == f2.director:
                FilmGraph.add_edge(f1,f2)
            for c in f1.cast:
                if c in f2.cast:
                    FilmGraph.add_edge(f1,f2)
```

```python
labels = nx.get_node_attributes(FilmGraph, 'title')
print(labels)
```

```python
options = {'node_color':'red', 'node_size':10, 'width':1, 'labels':labels, 'with_labels': True}
plt.figure(1, figsize = (30, 30))
nx.draw(FilmGraph, **options)
```
