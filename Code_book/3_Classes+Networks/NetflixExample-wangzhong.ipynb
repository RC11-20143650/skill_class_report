```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class Sanitaire:
    def __init__(self, Ligne, Station, Public, AccesBouton, Emplacement):
        self.Ligne = Ligne
        self.Station = Station
        self.Public = Public
        self.AccesBouton = AccesBouton
        self.Emplacement = Emplacement
```

```python
Sanitaires = []
with open("C:/Users/qq809/Desktop/SkillsClasses-main/3_Classes+Networks/sanitaires-reseau-ratp.csv",'r',encoding='mac_roman',newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Ligne = row['Ligne']
        Station = row['Station']
        Public = row['Public']
        AccesBouton = row['AccesBouton']
        Emplacement = row['Emplacement']
        Sanitaires.append(Sanitaire(Ligne, Station, Public, AccesBouton, Emplacement))
```

```python
print(Sanitaires[0].Ligne)
print(Sanitaires[0].Station)
```

```python
for c in Sanitaires[0].Ligne:
    print(c)
```

```python
SanitaireGraph = nx.Graph()
```

```python
SanitaireSubset = Sanitaires[0:48]
```

```python
for f in SanitaireSubset:
    SanitaireGraph.add_node(f)
```

```python
for f1 in SanitaireSubset:
    for f2 in SanitaireSubset:
        if not f1 == f2:
            if f1.Ligne == f2.Ligne:
                SanitaireGraph.add_edge(f1,f2)
            if f1.Emplacement == f2.Emplacement:
                SanitaireGraph.add_edge(f1,f2)
```

```python
labels = nx.get_node_attributes(SanitaireGraph, 'Station')
print(labels)
```

```python
options = {'node_color':'blue', 'node_size':1000, 'width':0.5,'edge_color':'blue', 'labels':labels, 'with_labels': True}
plt.figure(1, figsize = (30, 30))
nx.draw(SanitaireGraph, **options)
```

This graph shows the connections between sanitaires based their ligne.
